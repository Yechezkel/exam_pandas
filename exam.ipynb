```python
import pandas as pd
```

```python
# importing the interest rate in Israel 
url = r"BR.csv"
df = pd.read_csv(url)
il_df =  df.loc[ :  , ["TIME_PERIOD", "OBS_VALUE"]]
il_df["TIME_PERIOD"] = pd.to_datetime(il_df["TIME_PERIOD"])
il_df = il_df[il_df["TIME_PERIOD"].dt.day == 1]
il_df["year"] = il_df["TIME_PERIOD"].dt.year
il_df["month"] = il_df["TIME_PERIOD"].dt.month
il_df = il_df.drop(columns = ["TIME_PERIOD"])
il_df.rename(columns = {"OBS_VALUE": "il_interest_rate"}, inplace = True)
il_df = il_df.sort_values(by = ["year", "month"])
il_df.reset_index(drop = True ,inplace = True)
il_delta = [0]
for i in range(1, len(il_df)):
    il_delta.append(il_df.iloc[i]["il_interest_rate"] - il_df.iloc[i-1]["il_interest_rate"])
il_df["il_delta"] = il_delta
il_df
```

```python
#  importing the interest rate in USA
url = r"yield-curve-rates-1990-2023 (1).csv"
df = pd.read_csv(url)
us_df =  df.loc[ :  , ["Date", "1 Yr"]]
us_df["Date"] = pd.to_datetime(us_df["Date"], format='%m/%d/%y') 
us_df["year"] = us_df["Date"].dt.year
us_df["month"] = us_df["Date"].dt.month
us_df = us_df.drop_duplicates(subset=["month", "year"])
us_df = us_df.drop(columns = ["Date"])
us_df.rename(columns = {"1 Yr": "us_interest_rate"}, inplace = True)
us_df = us_df.sort_values(by = ["year", "month"], ascending = True)
us_df.reset_index(drop = True ,inplace = True)
us_delta = [0]
for i in range(1, len(us_df)):
    us_delta.append(us_df.iloc[i]["us_interest_rate"] - us_df.iloc[i-1]["us_interest_rate"])
us_df["us_delta"] = us_delta
us_df
```

```python
# importing the exchange rate between USD and ILS 
url = r"USD_ILS Historical Data (2).csv"
df = pd.read_csv(url)
ex_df =  df.loc[ :  , ["Date", "Price"]]
ex_df["Date"] = pd.to_datetime(ex_df["Date"])
ex_df = ex_df[ex_df["Date"].dt.day == 1]
ex_df["year"] = ex_df["Date"].dt.year
ex_df["month"] = ex_df["Date"].dt.month
ex_df = ex_df.drop(columns=["Date"])
ex_df.rename(columns={"Price": "usd_ils_exchange"}, inplace=True)
ex_df = ex_df.sort_values(by = ["year", "month"], ascending = True)
ex_df.reset_index(drop=True, inplace=True)
exchange_delta = [0]
for i in range(1, len(ex_df)):
    exchange_delta.append(ex_df.iloc[i]["usd_ils_exchange"] - ex_df.iloc[i-1]["usd_ils_exchange"])
ex_df["exchange_delta"] = exchange_delta
ex_df.head()
```
